# Implementation from Scratch

```python
import torch
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import torch.nn as nn

import pdb
from torch.autograd import Variable

import numpy as np

from matplotlib import pyplot as plt
from IPython import display
%matplotlib inline
display.set_matplotlib_formats('svg')

seed = 0
torch.manual_seed(seed) 
```

```python
def dropout_layer(x, drop_p):
#     pdb.set_trace()
    assert 0 <= drop_p <= 1
    select = torch.rand(x.shape[1])
    select = (select > drop_p).float()
    if drop_p == 1:
        return torch.zeros_like(x)
    else:
#         pdb.set_trace()
#         return torch.true_divide(x*select, 1-drop_p)
        return select*x / (1.0 - drop_p)

def dropout(X, drop_prob):
    assert 0 <= drop_prob <= 1
    # In this case, all elements are dropped out
    if drop_prob == 1:
        return torch.zeros_like(X)
    mask = (torch.FloatTensor(X.shape).uniform_(0, 1) > drop_prob).float()
    return mask * X / (1.0 - drop_prob)
```

```python
x = torch.arange(16).reshape(2, 8)
dropout_layer(x, 0.5), dropout_layer(x, 0), dropout_layer(x, 1)
```

## Read data

```python
batch_size = 256

trans = transforms.ToTensor()

trainset = datasets.FashionMNIST(root='../data/', transform=trans, train=True)
testset = datasets.FashionMNIST(root='../data/', transform=trans, train=False)

train_loader = DataLoader(dataset=trainset, batch_size=batch_size, shuffle=True, num_workers=4)
test_loader = DataLoader(dataset=testset, batch_size=batch_size, shuffle=True, num_workers=4)
```

## define model parameters

```python
dim_input, n_output, n_hidden1, n_hidden2 = 784, 10, 256, 256

w1 = Variable(torch.randn(dim_input, n_hidden1), requires_grad=True)
b1 = Variable(torch.zeros(n_hidden1), requires_grad=True)
w2 = Variable(torch.randn(n_hidden1, n_hidden2), requires_grad=True)
b2 = Variable(torch.zeros(n_hidden2), requires_grad=True)
w3 = Variable(torch.randn(n_hidden2, n_output), requires_grad=True)
b3 = Variable(torch.zeros(n_output), requires_grad=True)

# params = [w1, b1, w2, b2, w3, b3]
```

## define the model

```python
drop_p1, drop_p2 = 0.2,0.5

def net(x, train):
    x = x.reshape(x.shape[0], -1)
    relu = nn.ReLU()
#     pdb.set_trace()
    try:
        h1 = relu(torch.mm(x, w1) + b1)
    except:
        pdb.set_trace()
    if train:
        h1 = dropout_layer(h1, drop_p1)
    h2 = relu(torch.mm(h1, w2) + b2)
    if train:
        h2 = dropout_layer(h2, drop_p2)
    output = torch.mm(h2, w3) + b3
    
    return output
```

```python
def sgd(params, lr, bs):
    for param in params:
        param.data.sub_(lr*param.grad / bs)
        param.grad.data.zero_()
```

## Training & Testing

```python
def accuracy(y_hat, y):
#     pdb.set_trace()
    return (y == y_hat.argmax(dim=1)).float().mean().item()

def evaluate(net, loader):
    accs = []
    for idx, data in enumerate(loader):
        img, label = data
        output = net(img, False)
        accs.append(accuracy(output, label))
        
    return np.mean(accs)
```

```python
def train(epochs, lr):
    params = [w1, b1, w2, b2, w3, b3]
    loss_func = nn.CrossEntropyLoss()
#     optim = torch.optim.SGD()

    train_loss, train_acc, test_acc = [],[],[]
    for epo in range(epochs):
        epo_loss, epo_acc = [], []
#         net.train()
        for idx, data in enumerate(train_loader):
            img, label = data
            output = net(img, True)
            loss = loss_func(output, label)
            loss.backward()
            sgd(params ,lr, batch_size)
            
            epo_loss.append(loss.item())
            epo_acc.append(accuracy(output, label))
        
        train_loss.append(np.mean(epo_loss))
        train_acc.append(np.mean(epo_acc))
        test_acc.append(evaluate(net, test_loader))
        
        print(f'epoch: {epo}, train_loss: {train_loss[epo]}, train_acc: {train_acc[epo]}, test_acc: {test_acc[epo]}')
        
    return train_loss, train_acc, test_acc 
```

```python
train_loss, train_acc, test_acc = train(10, 0.5)
```

```python
def visual(train_loss, train_acc, test_acc):
#     pdb.set_trace()
    x = range(len(train_loss))
    plt.plot(x, train_loss, label='train_loss')
    plt.plot(x, train_acc, label='train_acc')
    plt.plot(x, test_acc, label='test_acc')
    
    plt.legend()
    plt.show()
```

```python
visual(train_loss, train_acc, test_acc)
```

# Concise Implementation

```python
def load_data(bs, num_workers):
    trans = transforms.ToTensor()
    train_data = torchvision.datasets.FashionMNIST(root='../data/', train=True, transform=trans)
    test_data = torchvision.datasets.FashionMNIST(root='../data/', train=False, transform=trans)
    
    train_loader = DataLoader(dataset=train_data, batch_size=bs, num_workers=num_workers, shuffle=True)
    test_loader = DataLoader(dataset=test_data, batch_size=bs, num_workers=num_workers, shuffle=True)
    
    return train_loader, test_loader
```

```python
def accuracy(y_hat, y):
#     pdb.set_trace()
    return (y == y_hat.argmax(dim=1)).float().mean().item()

def evaluate(net, loader):
    accs = []
    net.eval()
    for idx, data in enumerate(loader):
        img, label = data
        output = net(img.reshape(-1,784))
        accs.append(accuracy(output, label))
        
    return np.mean(accs)
        
```

```python
class DropoutModel(nn.Module):
    def __init__(self, dim_input=784, n_output=10, n_hidden1=256, n_hidden2=256, drop_p1=0.2, drop_p2=0.5):
        super(DropoutModel, self).__init__()
        self.net = nn.Sequential(nn.Linear(dim_input, n_hidden1, bias=True),
                                nn.ReLU(),
                                nn.Dropout(drop_p1),
                                nn.Linear(n_hidden1, n_hidden2, bias=True),
                                nn.ReLU(),
                                nn.Dropout(drop_p2),
                                nn.Linear(n_hidden2, n_output, bias=True))
        
    def forward(self, data):
        return self.net(data)
    
```

```python
def concise_train(epochs, lr, model=DropoutModel()):
    bs, num_workers = 256, 4
    train_loader, test_loader = load_data(bs, num_workers)
#     model = DropoutModel()
    print(net)
    loss = nn.CrossEntropyLoss()
    optim = torch.optim.SGD(model.parameters(), lr = lr)
    
    train_loss, train_acc, test_acc = [], [], []
    for epo in range(epochs):
        epo_loss, epo_acc = [], []
        model.train()
        for idx, data in enumerate(train_loader):
            img, label = data
#             pdb.set_trace()
            output = model(img.reshape(img.shape[0],784))
            optim.zero_grad()
            iter_loss = loss(output, label)
            iter_loss.backward()
            optim.step()

            epo_loss.append(iter_loss.item())
            epo_acc.append(accuracy(output, label))
            
        train_loss.append(np.mean(epo_loss))
        train_acc.append(np.mean(epo_acc))
        test_acc.append(evaluate(model, test_loader))
        
        print(f'epoch {epo} train loss: {train_loss[epo]}, train acc: {train_acc[epo]}, test acc: {test_acc[epo]}')

    return train_loss, train_acc, test_acc
```

```python
train_loss, train_acc, test_acc = concise_train(10, 0.2)
```

```python
visual(train_loss, train_acc, test_acc)
```
