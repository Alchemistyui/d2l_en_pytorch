## Activate Funcs

```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F

from matplotlib import pyplot as plt
from IPython import display
%matplotlib inline
display.set_matplotlib_formats('svg')
```

```python
def xyplot(x_vals,y_vals,name):
    x_vals=x_vals.detach().numpy() # we can't directly use var.numpy() because varibles might 
    y_vals=y_vals.detach().numpy() # already required grad.,thus using var.detach().numpy() 
    plt.plot(x_vals,y_vals) 
    plt.xlabel('x')
    plt.ylabel(name+'(x)')
```

### ReLU

```python
x = Variable(torch.arange(-8.0,8.0, 0.1,dtype=torch.float32), requires_grad=True)
relu = F.relu(x)
xyplot(x,relu,'ReLU')
```

```python
relu.backward(torch.ones_like(x))
xyplot(x,x.grad,'ReLU_derivative')
```

### Sigmoid Function

```python
x = Variable(torch.arange(-8.0,8.0, 0.1,dtype=torch.float32), requires_grad=True)
sig = torch.sigmoid(x)
xyplot(x,sig,'sigmoid')
```

```python
sig.backward(torch.ones_like(x))
xyplot(x,x.grad,'sigmoid_derivative')
```
