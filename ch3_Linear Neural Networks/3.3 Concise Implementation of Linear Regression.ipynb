## Generate dataset

```python
import torch
from torch.utils.data import Dataset,DataLoader
import torch.nn as nn

import numpy as np

from matplotlib import pyplot as plt
from IPython import display
%matplotlib inline
display.set_matplotlib_formats('svg')
```

```python
w = torch.tensor([2,-3.4]).reshape(2,1)
b = torch.tensor(4.2)
num_examples = 1000
batch_size = 8
epsilon = torch.randn(1000).reshape(-1, 1)

def synthetic_data(w, b, num_examples):
#     x = torch.zeros((num_examples, 2)).normal_()
    x = torch.randn([1000, 2])
    y = torch.mm(x, w) + b + epsilon
    return x, y
features, labels = synthetic_data(w, b, num_examples)
```

## Reading data

```python
class RegressionDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
        
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
    
    def __len__(self):
        return len(self.features)
    
```

## Define the model

```python
class LinearRegressionModel(nn.Module):
    def __init__(self, in_channel, out_channel):
        super(LinearRegressionModel, self).__init__()
        self.layer = nn.Sequential(
                        nn.Linear(in_channel, out_channel, bias=True))
        self.initialization()
        
    def forward(self, data):
        pred = self.layer(data)
        return pred
    
#     Initialize Model Parameters
# initializing parameters without giving input dimensions, but not preallocate enough space. 
# the initialization is deferred until the first time that we attempt to pass data through network. 
# since the parameters have not been initialized yet we cannot yet manipulate them in any way.
    def initialization(self):
        self.layer[0].weight.data = torch.Tensor(np.random.normal(size=(1,2),scale=0.01,loc=0))
        self.layer[0].bias.data = torch.zeros(1)
```

## Training

```python
def train(epochs):
    dataset = RegressionDataset(features, labels)
    loader = DataLoader(dataset=dataset,batch_size=batch_size,shuffle=True)

    net = LinearRegressionModel(2, 1)
    
    loss_func = nn.MSELoss(reduction='sum')
    optim = torch.optim.SGD(net.parameters(), lr=0.03)
    loss_hist = []
    
    for epo in range(epochs):
        for i, (data, label) in enumerate(loader):
            output = net(data)
            optim.zero_grad()
            loss = loss_func(output, label)
            loss.backward()
            optim.step()

            if i % 10 == 0:
                print(f'epoch {epo}, iter {i}: loss {loss.mean()}')
                loss_hist.append(loss.mean())
    return loss_hist, net
```

```python
loss_hist, net = train(1)
idx = list(range(len(loss_hist)))

plt.plot(idx, loss_hist)
plt.show()
```

## Eval

```python
def evaluate(params_prob, params):
    w_prob, b_prob = params_prob
    w, b = params
#     print(w_prob.squeeze().shape, w.shape)
    print(f'estimate param: {w_prob.squeeze(), b_prob}')
    print(f'true param: {w.squeeze(), b}')
    print(f'weight error: {w_prob.squeeze() - w.squeeze()}')
    print(f'bias error: {b_prob - b}')
    
w_prob = list(net.parameters())[0][0]
b_prob = list(net.parameters())[1][0]
evaluate([w_prob, b_prob], [w, b])
```
