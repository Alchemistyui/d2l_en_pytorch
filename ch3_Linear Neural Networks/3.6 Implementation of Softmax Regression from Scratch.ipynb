```python
import torch
from torch.utils.data import DataLoader
from torchvision import transforms
import torchvision
from torch.autograd import Variable

from IPython import display
from matplotlib import pyplot as plt
%matplotlib inline
display.set_matplotlib_formats('svg')

import numpy as np
import pdb
```

## load dataset

```python
trans = transforms.ToTensor()
train_data = torchvision.datasets.FashionMNIST(root='../data/', train=True, transform=trans)
test_data = torchvision.datasets.FashionMNIST(root='../data/', train=False, transform=trans)
```

```python
bs = 256
num_workers = 4
train_loader = DataLoader(dataset=train_data, shuffle=True, batch_size=bs, num_workers=num_workers)
test_loader = DataLoader(dataset=test_data, shuffle=True, batch_size=bs, num_workers=num_workers)
```

## Initialize model Params

```python
# img: 28*28 -> bs*784
# output: bs*10
# wx+b, weight: 784*10, bias: 1*10
input_channel = 784
output_channel = 10
w_hat = Variable(torch.zeros(input_channel, output_channel, dtype=torch.float32).data.normal_(std=0.01), requires_grad=True)
b_hat = Variable(torch.zeros(output_channel), requires_grad=True)
```

## Define the softmax operation

```python
def softmax(X):
    exp = torch.exp(X)
    exp_sum = exp.sum(axis=1, keepdims=True)
    return exp / exp_sum
```

```python
def test_softmax():
    X = np.random.normal(0, 1, (2, 5))
    X = torch.from_numpy(X)
    print(X)
    print(softmax(X))
    print(softmax(X).sum(1))
    
test_softmax()  
```

## Define the model

```python
def net(x):
#     pdb.set_trace()
    logit = torch.mm(x.reshape(-1, input_channel), w_hat) + b_hat
    return softmax(logit)
```

## loss func

! Here cross_entropy use another format but not its origin definition. [The reason is here](https://www.cnblogs.com/marsggbo/p/10401215.html)

equals to log_softmax and nll_loss two step.

nll_loss function receives two sensors as params, the first is output of the model, the second is the label target(index). Find the corresponding element in the output according to the value of the target, sum them together and take the average value.


```python
def my_loss(y, y_hat):
    assert y.shape[0] == y_hat.shape[0]
#     print(y.shape, y_hat.shape)
#     pdb.set_trace()
    loss = -((torch.log(y_hat.argmax(1).unsqueeze(1).float())*y).sum(axis=1, keepdims=True))
    loss.reqiures_grad=True
    return loss

def cross_entropy(y, y_hat):
    return -torch.gather(y_hat, 1, y.unsqueeze(dim=1)).log()
```

```python
y_hat = torch.tensor([[0.1, 0.3, 0.6], [0.3, 0.2, 0.5]])
y = torch.tensor([0, 2])
cross_entropy(y, y_hat) , my_loss(y.reshape(-1, 1), y_hat)
```

```python
def accuracy(y, y_hat):
    return (y == y_hat.argmax(dim=1)).float().mean().item()
```

```python
def evaluate(test_loader, net):
    accs = []
    for idx, data in  enumerate(test_loader):
        img, label = data
        output = net(img)
        acc = accuracy(label, output)
        accs.append(acc)
        
#     print(f'evaluate in test dataset accuracy: {np.mean(accs)}')
    return np.mean(accs)
    
```

```python
evaluate(test_loader, net)
```

## Optimization

```python
def sgd(params, lr, bs):
    for param in params:
        param.data.sub_(param.grad*lr/bs)
        param.grad.data.zero_()
```

## Training

```python
epochs, lr = 30, 0.3
def train():
    train_loss, train_acc, test_acc = [], [], []
    for epo in range(epochs):
        epo_loss, epo_acc = [],[]
        for idx, data in enumerate(train_loader):
            img, label = data
            output = net(img)
            loss = cross_entropy(label, output).mean()
            loss.backward()
            sgd([w_hat, b_hat], lr, bs)
            epo_loss.append(loss.item())
            epo_acc.append(accuracy(label, output))
            
        train_loss.append(np.mean(epo_loss))
        train_acc.append(np.mean(epo_acc))
        test_acc.append(evaluate(test_loader, net))
#         print(f'epoch {epo} train loss: {train_loss[epo]}, train acc: {train_acc[epo]}, test acc: {test_acc[epo]}')

    return train_loss, train_acc, test_acc
train_loss, train_acc, test_acc = train()
```

## Visualization

```python
def visualization(train_loss, train_acc, test_acc):
    x = list(range(len(train_loss)))
    plt.plot(x, train_loss, label='train_loss')
    plt.plot(x, train_acc, label='train_acc')
    plt.plot(x, test_acc, label='test_acc')
    
    plt.legend()
    
    plt.show()
    
visualization(train_loss, train_acc, test_acc)
```
