## 2.6.1 Basic Probability Theory

```python
import numpy as np
import torch
from torch.distributions.multinomial import Multinomial
import random
from matplotlib import pyplot as plt
import math
```

```python
# multinomial distribution
# for loop is very slow
random.randint(0,5)
times = torch.zeros(6)
for i in range(1000):
    times[random.randint(0,5)] += 1
    
print(times)
print(times/times.sum())
```

```python
# numpy ver
fair_probs = [1.0 /6 ] * 6
np.random.multinomial(10000, fair_probs)/10000
```

```python
# torch ver
probs = torch.ones(6) / 6
# total_count: run the sampler a bunch of times,
distribution = Multinomial(total_count=1, probs=probs)
distribution.sample()
```

```python
distribution = Multinomial(total_count=1000, probs=probs)
distribution.sample() / 1000
```

### how these probabilities converge over time towards the true probability.

```python
num_rolls = 1000
probs = torch.ones(6) / 6
distribution = Multinomial(total_count=1, probs = probs)
total = torch.zeros(6)
count = torch.zeros(6, 1000)
for i in range(num_rolls):
    total += distribution.sample()
    count[:, i] = total
total, count
```

```python
total / num_rolls
```

```python
plt.figure(figsize=[8,5])
for i in range(count.shape[0]):
    plt.plot(range(num_rolls), count[i]/(np.arange(1000)+1), label='die='+str(i))
    
plt.legend()
plt.title('converge of multinomial distribution')
plt.xlabel('sample num')
plt.ylabel('estimation of probability')
# plt.grid()

plt.axhline(y=0.16666, color='black', linestyle='dashed')

plt.show()
```

## The categorical distribution

```python
times = 10000
y = np.random.uniform(0,1,times)
x = np.arange(1,times+1)

# Count number of occurrences and divide by the number of total draws
p0 = np.cumsum(y < 0.35) / x
p1 = np.cumsum(y >= 0.35) / x

plt.figure(figsize=(11, 6))
plt.semilogx(x, p0)
plt.semilogx(x, p1)
plt.show()
```

```python
# another example
# p=[0.1,0.2,0.05,0.3,0.25,0.1] 
times = 10000

y = np.random.uniform(0,1, times)
x = np.arange(1, times+1)
probs = [0.1,0.2,0.05,0.3,0.25,0.1] 
distribution = []
for i in range(len(probs)):
    distribution.append(np.cumsum(y<probs[i]) / x)
# p0 = np.cumsum(y<probs[0]) / x
# p0
# len(distribution), distribution[0].shape

plt.figure(figsize=(10,6))

for i in range(len(probs)):
    plt.plot(x, distribution[i], label='prob No.'+str(i+1))
    
plt.legend()
plt.show()
```

## Normal distribution

```python
x = torch.arange(-10, 10, 0.1)
p = (1/math.sqrt(2*np.pi)) * np.exp(-0.5*x**2)
plt.plot(x, p)
plt.show()
```

The normal distribution has yet another desirable property. In a way all distributions converge to it, if we only average over a sufficiently large number of draws from any other distribution.

## Central Limit Theorem

It states that for sufficiently well-behaved random variables, in particular random variables with well-defined mean and variance, the sum tends toward a normal distribution.

```python
# Generate 10 random sequences of 10,000 uniformly distributed random variables
tmp = np.random.uniform(size=(10000,10))
x = 1.0 * (tmp > 0.3) + 1.0 * (tmp > 0.8)
mean = 1 * 0.5 + 2 * 0.2
variance = 1 * 0.5 + 4 * 0.2 - mean**2
print('mean {}, variance {}'.format(mean, variance))

# Cumulative sum and normalization
y = np.arange(1,10001).reshape(10000,1)
z = np.cumsum(x,axis=0) / y

plt.figure(figsize=(10,5))
for i in range(10):
    plt.semilogx(y,z[:,i])

plt.semilogx(y,(variance**0.5) * np.power(y,-0.5) + mean,'r')
plt.semilogx(y,-(variance**0.5) * np.power(y,-0.5) + mean,'r')
plt.show()
```
